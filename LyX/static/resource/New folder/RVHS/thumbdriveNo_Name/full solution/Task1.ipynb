```python
# Task 1.1
# Write a function task1_1(name_A, name_B) where name_A and name_B are strings 
# which consists of alphabet letters and spaces. The function will return True if 
# A) The alphabet letters used in string name_A and name_B are the same and
# B) the spaces in string name_A and name_B are at the same locations.
# For example,
# >>> match_names("Abcde", "Deabc")
# True
# >>> match_names("Abcde Fgh I", "Ihgfe Dcb A")
# True
# >>> match_names("Abcd Efgh I", "Ihgfe Dcb A")
# False
# >>> match_names("Abcde Fzh I", "Ihgfe Dcb A")
# False

import random

def task1_1(name_A, name_B):
    #lenght check [1]
    if len(name_A) != len(name_B):
        return False
    #check for space positions [2]
    for i in range(len(name_A)):
        if name_A[i] == ' ':
            if name_B[i] != ' ':
                return False
        if name_B[i] == ' ':
            if name_A[i] != ' ':
                return False
    #Check for each letter [3]
    letter_dict_A = {}
    for char in name_A:
        if char.upper() not in letter_dict_A:
            letter_dict_A[char.upper()] = 1
        else:
            letter_dict_A[char.upper()] += 1
            
    letter_dict_B = {}
    for char in name_B:
        if char.upper() not in letter_dict_B:
            letter_dict_B[char.upper()] = 1
        else:
            letter_dict_B[char.upper()] += 1
    
    for key in letter_dict_A.keys():
        if key not in letter_dict_B:
            return False
        if letter_dict_B[key.upper()] != letter_dict_A[key.upper()]:
            return False
    return True #correctly return results [1]

def quick_test():
    print(task1_1("Abcde", "Deabc")) #True
    print(task1_1("Abcde Fgh I", "Ihgfe Dcb A")) #True
    print(task1_1("Abcd Efgh I", "Ihgfe Dcb A")) #False
    print(task1_1("Abcde Fzh I", "Ihgfe Dcb A")) #False
quick_test()
    
def random_test():
    def rearrange(name):
        result = ""
        name = name.replace("'","")
        name = name.replace("-","")
        name_lst = list(name.replace(" ",""))
        random.shuffle(name_lst)

        words = name.split(" ")
        curr = 0
        for i in range(len(words)):
            no_char = len(words[i])
            isUpper = True
            for j in range(no_char):
                if isUpper:
                    result += name_lst[curr].upper()
                    isUpper = False
                else:
                    result += name_lst[curr].lower()
                curr += 1
            result += " "
        return result[:-1]
    for i in range(10):
        available_chars = "ABCDEFGHIJKLMNOPQRSTUVWXYZ    "
        random_name = ""
        name_length = random.randint(10, 30)
        for i in range(name_length):
            random_name += available_chars[random.randint(0, len(available_chars)-1)]
        random_name = random_name.strip()
        new_form = rearrange(random_name)
        print(task1_1(random_name, new_form), random_name + " / " + new_form)

random_test()
```

```python
def readfile(filename):
    result = []
    f = open(filename, 'r')
    line = f.readline()
    while line:
        result.append(line[:-1])
        line = f.readline()
    f.close()
    return result

def task1_2():
    # O(n) / O(n**2) [5/1]
    # readfile [2]
    # Able to find match [3]
    # Identify 7 match cannot be found [2]
    # 1 collision resolution resolved correctly [1]
    def hashname(name):
        result = 0
        temp = name.upper()
        for i in range(len(temp)):
            if temp[i] == ' ':
                result += i*ord(temp[i])*10
            else:
                result += ord(temp[i])
            
        return result
    
    lstA = readfile("namelist_A.txt")
    lstB = readfile("namelist_B.txt")    
    h1 = {}
    for i in range(len(lstB)):
        if hashname(lstB[i]) not in h1:
            h1[hashname(lstB[i])] = [lstB[i]]
        else:
            h1[hashname(lstB[i])].append(lstB[i])
    
    print("{:40}{:40}".format("From namelist_A.txt", "From namelist_B.txt"))
    for i in range(len(lstA)):
        if hashname(lstA[i]) not in h1:
            print("{:40}{:40}".format(lstA[i], "***********No match***********"))
        elif len(h1[hashname(lstA[i])]) == 1:
            print("{:40}{:40}".format(lstA[i], h1[hashname(lstA[i])][0]))
        else:
            # assuming the names in match_name_A always has a unique pair in match_name_B
            for name in h1[hashname(lstA[i])]:
                if task1_1(lstA[i], name):
                    print("{:40}{:40}".format(lstB[i], name))
            

task1_2()
```

```python
def merge(lst1, lst2):
    lst = []
    def merge_helper(lst1, lst2):
        if len(lst1) == 0:
            lst.extend(lst2)
        elif len(lst2) == 0:
            lst.extend(lst1)
        else:
            if lst1[0] > lst2[0]:
                lst.append(lst1.pop(0))
            else:
                lst.append(lst2.pop(0))
            merge_helper(lst1, lst2)
    merge_helper(lst1, lst2)
    return lst

def msort(lst):
    if len(lst) == 1:
        return lst
    else:
        mdpt = len(lst)//2
        return merge(msort(lst[:mdpt]), msort(lst[mdpt:]))
    
print("----Merge sort-----")
lst = [6,1,4,8,5,9,7]
print(msort(lst))



def partition(lst):
    def helper(lst, solution, pivot, pivot_ptr):
        if lst:
            if lst[0] < pivot:
                return helper(lst[1:], [lst[0],] + solution, pivot, pivot_ptr + 1)
            else:
                return helper(lst[1:], solution + [lst[0],], pivot, pivot_ptr)
        else:
            return solution, pivot_ptr
    return helper(lst[1:], [lst[0],], lst[0], 0)



def qsort(lst):
    if lst:
        partitioned_lst, pivot_ptr = partition(lst)
        return qsort(partitioned_lst[:pivot_ptr]) + [partitioned_lst[pivot_ptr],] + qsort(partitioned_lst[pivot_ptr + 1:])
    else:
        return []

print("----Quick sort-----")
lst = [5,1,7,2,8,3,9,13,20,1]
lst = [100,5,1,7,2,8,3,9,13,20]
#print(partition(lst))
print(qsort(lst))
```
