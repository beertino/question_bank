```python
import pymongo, csv, json

client = pymongo.MongoClient("127.0.0.1", 27017)

client.drop_database("products")

db = client["products"]
coll = db["balloons"]

with open('balloons.json') as file:
    data = json.load(file)

coll.insert_many(data)

print("Task 4.1")
print()
result = coll.find()
for doc in result:
    print(doc)

print()
print()

result = coll.find({'design': 'car'})
print("Task 4.2")
print()
print([item['amount'] for item in result])
    
print()

search = {'$or': [{'helium': None}, {'helium':{'$exists':False} } ]}
update = {'$set':{'helium':"no"}}

print()
print()

coll.update_many(search, update)
result = coll.find()
print("Task 4.3")
print()
for doc in result:
    print(doc)
    
print()
print()


result = coll.find({'$and': [{'helium': "no"},{'$or':[{'colours':{'$in':['green']}},{'colours':{'$nin':['black']}}] } ]})
print("Task 4.4")
print()
print([item['design'] for item in result])

```
