```python
# Breaking up the whole paper into individual questions
def break_qpaper(school_name:str,year:str,paper_type:str):

    syllabus_dict = {'2021':'9569', '2020':'9569', '2019':'9597','2018':'9597','2017':'9597','2016':'9597','2015':'9597','2014':'9597','2013':'9597'}

    with open(f'./{paper_type}/{year}/{syllabus_dict[year]}-{school_name}-{year}.tex') as f:
        content = f.read()

    # The string 'SPLIT_HERE' is used as a separator between questions
    qn_list = content.split('{[}SPLIT\_HERE{]}')
    qn_list = [qn.strip() for qn in qn_list]

    first = 1

    # Based on where the separator 'SPLIT_HERE' is located
    # the first question would contain the string '\begin{enumerate}'
    # which should be purged
    qn_list[first] = ''.join(list(qn_list[1])[17:]).strip()

    # Writing the questions in separate file
    for i in range(1,len(qn_list)-1):

        # extracting the metadata in the string and 
        # get the paper number and question
        metadata = qn_list[i]
        start = metadata.index('{[}') + 2 
        end = metadata.index('{]}')
        metadata = metadata[start:end].split('/')
        paper_number = metadata[-2]
        question_number = metadata[-1]
        
        # writing the questions
        with open(f'./indiv_qns/{year}_{school_name}_{paper_type}_{paper_number}_{question_number}.tex','w') as f:
            f.write(qn_list[i])
```

```python
years = ['2021','2020','2019','2018','2017','2016','2015','2014','2013']
paper_type = 'PRELIMS'

school_names = ['ACJC','DHS','HCI','JPJC','NYJC','RVHS','YIJC','JJC','IJC','PJC','ALVL']

for year in years:
    for school in school_names:
        try:
            break_qpaper(school,year,paper_type)
        except:
            print(f'({school},{year},{paper_type}) is not a valid parameter.')
```
