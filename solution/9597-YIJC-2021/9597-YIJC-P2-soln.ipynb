```python
# Task 1.3

def insert_element(arr,elem):
    i = 0
    while elem.get_tag() <= arr[-1-i].get_tag():
        i +=1 
        if i == len(arr):
            return [elem]+arr
    if i == 0:
        return arr+[elem]
    else:
        return arr[:-i]+[elem]+arr[-i:]

def insertionSort(arr):
    sorted_subarr=[arr[0]]
    for i in range(1,len(arr)):
        if arr[i].get_tag()>sorted_subarr[-1].get_tag():
            sorted_subarr.append(arr[i])
        else:
            sorted_subarr=insert_element(sorted_subarr,arr[i])
    return sorted_subarr

# Checker lines
Sorted_PersonList= insertionSort(PersonList)
for i in Sorted_PersonList:
    print(i.get_tag(),i.get_name())
```

```python
# Task 1.4
count = 0

def search(lst, tag):
    global count
    if len(lst) == 0:
        return None
    if len(lst) == 1 :
        # the single element in the list might or might not be of the correct tag
        # when the tag is wrong it'll return the element at extreme ends of the list but with the wrong tag
        return lst[0] if lst[0].get_tag()==tag else None
    else:
         mid = len(lst)//2
         print(mid,[i.get_name() for i in lst])
         if tag == lst[mid].get_tag():
             count += 1
             return lst[mid]
         elif tag < lst[mid].get_tag():
             count += 1
             return search(lst[:mid],tag)
         elif tag > lst[mid].get_tag():
             count += 1
             return search(lst[mid+1:],tag)

print(search(Sorted_PersonList,957126))
print(count)
```

```python
# Task 4.3
import csv

class Node:
    def __init__(self,Name, Exp,Pointer=None):
        self.Name = Name
        self.Exp = Exp
        self.Pointer = Pointer

    def setPointer(self,pointer):
        self.Pointer = pointer
    
    def getPointer(self):
        return self.Pointer
    
    def setName(self,name):
        self.name = name
    
    def getName(self):
        return self.Name

    def setExp(self,exp):
        self.Exp = exp

    def getExp(self):
        return self.Exp

class StudentList:
    def __init__(self):
        self.Start = None

    # The Add here refers to appending the elements to the end
    def Add(self, node):
        if self.Start == None:
            self.Start = node
        else:
            last = self.Start
            while last.Pointer:
                last = last.Pointer

            last.Pointer = node

    def Display(self):
        current_node = self.Start
        while current_node:
            print(current_node.getName(),current_node.getExp())
            current_node = current_node.Pointer

    def Update(self,name,points):
        current_node = self.Start
        while current_node:
            if current_node.getName() == name:
                current_node.setExp(int(current_node.getExp())+points)
                print(f"{current_node.getName()}'s exp is now {current_node.getExp()}")
                return current_node
            current_node = current_node.Pointer

    def Delete(self, name):
        prev_node = None
        current_node = self.Start

        # the following loop only starts if self.Start is not None, i.e. when the list is nonempty
        # the method will then return False straight up
        while self.Start:
            if current_node.getName() == name:
                # the following if block will run if prev_node is not None, meaning that we're not
                # at the start of the list. So we just get the prev_node to point next node after current_node
                if prev_node:
                    prev_node.setPointer(current_node.getPointer())
                # now, in this case, we're at the start of the list, so we just change our starting node in the list
                # to the next node after current_node
                else:
                    self.Start = current_node.getPointer()
                return True
            
            #we just traverse through the nodes here. 
            prev_node = current_node
            current_node = current_node.getPointer()

with open('DATA.txt') as f:
    reader = csv.reader(f)
    reader = list(reader)

new_student_list = StudentList()
for row in reader:
    new_node = Node(row[0],row[1])
    new_student_list.Add(new_node)


new_student_list.Delete('YZEMIN')

new_student_list.Display()
```
