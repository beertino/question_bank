```python
# Task 2.1 (3 marks)

class Node:
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None
    
    def set_left(self,left):
        self.left = left
    
    def set_right(self,right):
        self.right = right
    
    def get_data(self):
        return self.data
    
    def get_left(self):
        return self.left
    
    def get_right(self):
        return self.right


# Task 2.2 (11 marks)
filename = input("Please enter filename: ")
f1 = open(filename)

freq_dict = {}
inputstring = ''

for line in f1:
    for char in line:
        inputstring += char
        if char in freq_dict:
            freq_dict[char] = freq_dict[char]+1
        else:
            freq_dict[char] = 1
f1.close()

freq_list = []

for char in freq_dict:
    node = Node(char)
    index = 0
    for item in freq_list:
        if freq_dict[item.get_data()] < freq_dict[char]:
            index += 1
    freq_list.insert(index, node)

    
# Task 2.3 (5 marks)

index = 0
while index < len(freq_list)-1:
    node1 = freq_list[index]
    node2 = freq_list[index+1]
    
    new_data = node1.get_data()+node2.get_data()
    new_freq = freq_dict[node1.get_data()] + freq_dict[node2.get_data()]
    
    freq_dict[new_data] = new_freq
    
    new_node = Node(new_data)
    new_node.set_left(node1)
    new_node.set_right(node2)
    
    insert_index = 0
    for item in freq_list:
        if freq_dict[item.get_data()] < new_freq:
            insert_index += 1
    freq_list.insert(insert_index, new_node)
    
    index += 2

# Task 2.4 (8 marks)

def get_path(node,char):
    if char in node.get_data():
        if node.get_data() == char:
            return ''
        elif char in node.left.get_data():
            return '0'+get_path(node.get_left(),char)
        else: # char in node.right.get_data()
            return '1'+get_path(node.get_right(),char)
    else:
        return None

convert_dict = {}
root = freq_list[-1]

for string in freq_dict:
    if len(string) == 1:
        convert_dict[string] = get_path(root,string)

outputstring = ''
for char in inputstring:
    outputstring += convert_dict[char]
    
f2 = open(filename[-4]+"_compressed.txt",'w')
f2.write(outputstring)
f2.close()
```
