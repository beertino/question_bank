```python
# Task 3.1
# Write program code to read the csv file "health_facilities.csv" and insert all information in the file as documents
# into a NoSQL MongoDB database called "Health" with one collection called "facilities".
# The "_id" of the documents in the database should start from 1, 2, 3 and 4 etc. 
# The correct data type of each field is expected to be inserted into the database.

from pymongo import MongoClient
import csv
# [3(setup) + 1(csv) + 1(skip 1st) + 1(_id increment) + 3(create dict) + 1(insert_one)]
client = MongoClient('localhost', 27017)
client.drop_database('Health')
mydb = client['Health']
mycol = client['Health']['facilities']

with open('health_facilities.csv', mode ='r')as file:
    # reading the CSV file
    csvFile = csv.reader(file)
 
    # displaying the contents of the CSV file
    results = []
    for line in csvFile:
        results.append(line)
    
    for row in range(1, len(results)):
        post = {"_id": row}
        for col in range(len(results[0])):
            if results[row][col].isdigit():
                post[results[0][col]] = int(results[row][col])
            else:
                post[results[0][col]] = results[row][col]
        print(post)
        x = mycol.insert_one(post)

#client.close()
```

```python
# Task 3.2 
# a) Write a MongoDB Pymongo query to retrieve public acute hospital documents with
# their corresponding number of beds more than 7200. #[4]
query = {"$and":[{"institution_type":"Hospital"},\
                 {"facility_type_a":"Acute"},\
                 {"public_private":"Public"},\
                 {"no_beds":{"$gt":7200}}]}
posts = mycol.find(query)

# b) Write program code to bubble sort the results retrieved in Task 3.2 a) according to
# the average number of beds per facility. Then, display the top 3 years which has the 
# highest average number of beds per facility using the format below.
# "The three years that have the lowest average number of beds per facility are: _____, _____ and _____."


# [1(average) + 5(bubble) + 1(top 3)]
results = []

for post in posts:
    results.append([post["_id"], post["year"], post["no_beds"]/post["no_of_facilities"]])

def bubble_sort(lst):
    for i in range(len(lst)-1, 0, -1):
        for j in range(i):
            if lst[j][2] > lst[j+1][2]:
                lst[j+1],lst[j] = lst[j],lst[j+1]

bubble_sort(results)
print("The three years that have the lowest average number of beds per facility are: {}, {} and {}".format(str(results[0][1]),str(results[1][1]),str(results[2][1])))

```
